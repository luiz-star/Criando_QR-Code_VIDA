## Geração de QR Code usando Python

Código QR significa Código de Resposta Rápida. Os códigos QR podem parecer simples, mas são capazes de armazenar muitos dados. Independentemente da quantidade de dados que eles contêm quando o código QR digitalizado permite que o usuário acesse as informações instantaneamente. É por isso que eles são chamados de Código de Resposta Rápida.

Estes estão sendo usados ​​em muitos cenários nos dias de hoje. Ele apareceu pela primeira vez no Japão em 1994. Os códigos QR podem ser usados ​​para armazenar (codificar) muitos dados e também de vários tipos. Por exemplo, eles podem ser usados ​​para codificar:

* Detalhes do contato
* IDs do Facebook, IDs do Instagram, IDs do Twitter, IDs do WhatsApp e muito mais.
* detalhes do evento
* Links do Youtube
* Detalhes do produto
* Link diretamente para baixar um aplicativo na Apple App Store ou Google Play.
* Eles também estão sendo usados ​​para fazer transações digitais simplesmente digitalizando códigos QR.
* Acesse o Wi-Fi armazenando detalhes de criptografia, como SSID, senha e tipo de criptografia.

Acabamos de ver algumas vantagens dos códigos QR. Agora vamos aprender aqui como podemos gerar códigos QR em Python.

Para geração de código QR usando python, usaremos um módulo python chamado QRcode.

Link: https://pypi.org/project/qrcode/

Instale-o usando este comando: pip install qrcode

Iremos gerar um QR Code para codificar o link do youtube e também exploraremos mais. A geração de código QR é simples. Basta passar o texto, link ou qualquer conteúdo para a função 'fazer' do módulo QRcode.

```python
!pip install qrcode
```

```python
# Importando as Bibliotecas

import qrcode
img = qrcode.make("https://www.youtube.com/")
img.save("youtubeQR.jpg")
```

```python
# Saída do código acima
img
```

Você pode digitalizá-lo e verificar.( Utilize o leitor de QR Code de seu celular por exemplo e leia a imagem acima)

Você pode ver que são apenas 3 linhas de código para gerar este QR Code. Mais uma coisa a mencionar é que não é necessário que você sempre tenha que fornecer um link para a função qrcode.make(). Você também pode fornecer texto simples.

Por exemplo: VIDA = Virtual Intelligence of Data Academy

```python
import qrcode
img = qrcode.make("VIDA = Virtual Intelligence of Data Academy")
img.save("youtubeQR.jpg")
img
```

Digitalize-o do seu celular e você receberá o conteúdo.

Portanto, esta é a única parte, que envolve gerar um QR Code e digitalizá-lo. Mas e se quisermos ler este QR Code, ou seja, agora queremos saber o que foi codificado no QR Code sem escaneá-lo. Para isso, usaremos o OpenCV. 

OpenCV é uma biblioteca de funções de programação focada em tarefas de visão computacional em tempo real.

Instale o opencv: pip install opencv-python

Código para decodificar um código QR de volta para conhecer a string original.

```python
!pip install opencv-python
```

```python
import cv2
d = cv2.QRCodeDetector()
val, _, _ = d.detectAndDecode(cv2.imread("youtubeQR.jpg"))
print("Decoded text is: ", val)
```

## OpenCV-Python é uma biblioteca de ligações Python projetadas para resolver problemas de visão computacional . ...
